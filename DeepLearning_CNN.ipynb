```
Dir = "cats_and_dogs_filtered"
train_dir = Dir + "/train"
test_dir = Dir + "/validation"

train_cat_dir = train_dir + "/cats"
train_dog_dir = train_dir + "/dogs"


test_cat_dir = test_dir + "/cats"
test_dog_dir = test_dir + "/dogs"
```

```
import os
```

```
###reading images:-
train_cat_images = os.listdir(train_cat_dir)
train_dog_images = os.listdir(train_dog_dir)
```

```
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
### creating our own CNN model and layers:-
def Create_Model(h,w,d):
  model = Sequential()
  model.add(Conv2D(32, (3,3), activation="relu", input_shape = (h,w,d)))
  model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2,2)))
  model.add(Dropout(0.25))

  model.add(Conv2D(64, (3,3), activation="relu", input_shape = (h,w,d)))
  model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2,2)))
  model.add(Dropout(0.25))

  model.add(Conv2D(128, (3,3), activation="relu", input_shape = (h,w,d)))
  model.add(BatchNormalization())
  model.add(MaxPooling2D(pool_size=(2,2)))
  model.add(Dropout(0.25))

  model.add(Flatten())
  model.add(Dense(512, activation="relu"))
  model.add(BatchNormalization())
  model.add(Dropout(0.5))

  model.add(Dense(2, activation="softmax"))
  model.compile(loss="categorical_crossentropy", optimizer="RMSprop", metrics=["accuracy"])
  return model


```

```
### passing images to to the created cnn model
### regenerating the images
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
data_generator = ImageDataGenerator(1.0/255)
train_data_gen = data_generator.flow_from_directory(train_dir, target_size=(150,150), batch_size=20, class_mode="categorical")

test_data_gen = data_generator.flow_from_directory(test_dir, target_size=(150,150), batch_size=20, class_mode="categorical")
```

```
a = Create_Model(150,150,3)
a
```

```
history = a.fit(train_data_gen, epochs=20, validation_data=test_data_gen)
```
