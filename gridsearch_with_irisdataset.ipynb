```
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```
### grid search with epochs and batch_size:-
def Create_Model():
    model = Sequential()
    model.add(Dense(12, input_shape = (4,), activation = "relu"))
    model.add(Dense(3, activation = "softmax"))
    model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
    return model


iris = load_iris()
x = iris.data
y = iris.target

scaller = StandardScaler()
x = scaller.fit_transform(x)
y = y.reshape(-1, 1)
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()

model = KerasClassifier(build_fn=Create_Model)
batch_size = [20, 40, 80]
epochs = [10, 30, 50, 70, 100]

param_grid = dict(batch_size = batch_size, epochs = epochs)

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = 3, n_jobs=1)
result = grid.fit(x, y)
```

```
print("Best: {} using {}".format(result.best_score_, result.best_params_))
```

```
### grid search with optimizer:-
def Create_Model(optimizer = "adam"):
    model = Sequential()
    model.add(Dense(12, input_shape = (4,), activation = "relu"))
    model.add(Dense(3, activation = "softmax"))
    model.compile(loss = "categorical_crossentropy", optimizer = optimizer, metrics = ["accuracy"])
    return model


iris = load_iris()
x = iris.data
y = iris.target

scaller = StandardScaler()
x = scaller.fit_transform(x)
y = y.reshape(-1, 1)
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()

model = KerasClassifier(build_fn=Create_Model, verbose = 0)
opt = ["sgd", "adam", "RMSProp", "Adagrad", "Adamax"]
param_grid = dict(optimizer = opt)

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = 3, n_jobs=1)
result = grid.fit(x, y)
```

```
print("Best: {} using {}".format(result.best_score_, result.best_params_))
```
